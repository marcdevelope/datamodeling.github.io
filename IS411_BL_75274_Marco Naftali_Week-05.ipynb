2. Persebaran Usia untuk Setiap Peringkat:Melihat kolom tabel diatas bisa mengevaluasi sebaran usia untuk setiap peringkat. Ini dapat memberikan gambaran tentang demografi usia pengguna yang memberikan peringkat tertentu.

Contoh wawasan : 
- Pengguna dengan usia di atas rata-rata memberikan peringkat lebih tinggi atau lebih rendah.

- Dapat terlihat pola tertentu dalam preferensi peringkat tergantung pada kelompok usia.

```python
#17
character_gender_info = review.groupby(['favourite_character', 'user_sex']).agg({'user_age': 'mean', 'user_sex': 'count'})
character_gender_info.columns = ['average_age', 'gender_count']
print(character_gender_info)
```

```python
#18 
random_samples = review.sample(n=5, random_state=42)
random_samples
```

```python
#19 
review.groupby('favourite_character', group_keys=False).apply(lambda x: x.sample(frac=0.1))
```

```python
#20 
plt.figure(figsize=(12, 6))
sns.barplot(x='favourite_character', y='user_age', hue='user_sex', data=review)
plt.xlabel('Favorit Character')
plt.ylabel('Umur User')
plt.title('Rata-rata Distribusi antara favorit dengan umur user')
plt.show()
```
